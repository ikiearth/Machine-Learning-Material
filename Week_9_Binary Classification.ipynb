```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/ikiearth/Binary-Clasification/main/binary_classification_sand_production.csv')
dataset
```

```python
dataset.describe()
```

```python
dataset['SAND'].value_counts()
```

```python
X = dataset.drop(['SAND', 'No'], axis=1)
y = dataset['SAND']    
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=1)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_train = pd.DataFrame(X_train, columns=X.columns)
X_train
```

```python
X_test = pd.DataFrame(X_test, columns=X.columns)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred)).plot()
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
kn = KNeighborsClassifier(n_neighbors=10, weights= 'distance')
kn.fit(X_train, y_train)
```

```python
y_pred_kn = kn.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_kn))
```
