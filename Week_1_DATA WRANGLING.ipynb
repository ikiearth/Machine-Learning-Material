```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Collection

```python
confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
recovery = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')

print(confirmed.shape, recovery.shape, deaths.shape)
confirmed.head(10)
#recovery.head(10)
#deaths.head(10)
```

# Hapus Kolom yang Tidak Digunakan dan Rename
- Province/state
- Lat
- Long

```python
tables = (confirmed, recovery, deaths)

for table in tables:
    table.drop(['Province/State', 'Lat', 'Long'], axis = 1, inplace = True)
    table.rename({'Country/Region' : 'Country'}, axis = 1, inplace = True)

confirmed.head(10)
```

# Menyatukan Data atau Aggregation

```python
confirmed = confirmed.groupby('Country').sum()
recovery = recovery.groupby('Country').sum()
deaths = deaths.groupby('Country').sum()

confirmed.head(10)
```

```python
for table in tables:
    print(table.shape, "\n")
```

# Transpose

```python
confirmed = confirmed.stack().reset_index().rename({'level_1': 'Date', 0: 'Confirmed'}, axis=1)
recovery = recovery.stack().reset_index().rename({'level_1': 'Date', 0: 'Recovery'}, axis=1)
deaths = deaths.stack().reset_index().rename({'level_1': 'Date', 0: 'Deaths'}, axis=1)

tables = (confirmed, recovery, deaths)
confirmed.head(10)
```

# Merge Data

```python
df0= pd.merge(confirmed, recovery, how='left', left_on=['Country', 'Date'], right_on=['Country', 'Date'])
df = pd.merge(df0, deaths, how='left', left_on=['Country', 'Date'], right_on=['Country', 'Date'])
df['Date'] = pd.to_datetime(df.Date)
print(df)
```

# Add New Active Cases

```python
df.isna().sum()
```

```python
# Active Cases = confirmed - (recovery + deadths)

df['Active Cases'] = df['Confirmed'] - (df['Recovery'] + df['Deaths'])
print(df.head(100))
```

# Add New Date Rate

```python
df.isna().sum()
```

# Kasus Kenaikan Harian

```python
#kasus haru ke x - kasus hari ke (x-1) = kenaikan

df = df.sort_values(['Country', 'Date'])
df['Increase'] = df.groupby('Country')['Confirmed'].diff().fillna(0).astype(int)
print(df.head(200))
```

# Kasus Pertama Ditemukan

```python
firstday = df[df['Confirmed'] >= 1].groupby('Country').min().reset_index()[['Country', 'Date']]
firstday.rename({'Date':'FirstDays'}, axis= 1, inplace= True)
df = pd.merge(df, firstday, how= 'left', left_on=['Country'], right_on= ['Country'])

print(df.head(10))
```

```python
print(firstday.head(100))
```

# Membuat field Hari Ke-X dari Kasus Pertama Kali Muncul

```python
df['day_x'] = ((df['Date'] - df['FirstDays']).astype(str))
funct = lambda x : int(x[0:-5])

df['day_x'] = (df['day_x'].apply(funct)) + 1
df.loc[df['day_x'] < 0, ['day_x']] == 0

print(df.loc[34])
```

# Export to CSV

```python
df.to_csv('Covid19 Dataset.csv', index=None)
```
