### Learn from ChatGPT: 3 Input, 3 Output

```python
data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/[2022-2] Kecerdasan Buatan di Industri Minyak dan Gas/for Student (shared)/Chapter6_Frac_Stage_Data.xlsx')
data.drop(['Time'], axis=1, inplace=True)
```

```python
i=1
length=60
data.to_numpy()[i:i+length].shape
```

```python
# create input/output sequences for the RNN
length = 60
X = []  # input array
y = []  # target array
for i in range(len(data) - length - 1):
  X.append(data.to_numpy()[i:i+length])
  y.append(data.to_numpy()[i+length])
```

```python
# convert list to array
X = np.array(X)
y = np.array(y)

X.shape, y.shape
```

```python
# split the data into training and testing sets
split_index = 4400
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
n_features = X_train.shape[2]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN

# Build the RNN model
model = Sequential()
model.add(SimpleRNN(units=60, activation='relu', input_shape=(length, n_features)))
model.add(Dense(units=3))

# Compile and train the model
model.compile(optimizer='adam', loss='mse')
history = model.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
# Make predictions
y_train_pred = model.predict(X_train)
```

```python
plt.figure(figsize=(10,10))
for i in range(n_features):
  plt.subplot(n_features, 1, i+1)
  plt.plot(y_train_pred[:,i], label='Prediction')
  plt.plot(y_train[:,i], label='Actual')
  plt.title(data.columns[i])
  plt.legend()
  plt.tight_layout()
```

```python
y_test_pred = model.predict(X_test)
```

```python
# prepare x-axis for testing set
x_min = X_train.shape[0]
x_max = X_train.shape[0] + X_test.shape[0]

plt.figure(figsize=(10,10))
for i in range(n_features):
  plt.subplot(n_features, 1, i+1)
  plt.plot(np.arange(x_min, x_max), y_test_pred[:,i], label='Prediction (Test)')
  plt.plot(np.arange(x_min, x_max), y_test[:,i], label='Actual (Test)')
  # plt.plot(y_train_pred[:,i], label='Prediction (Train)')
  # plt.plot(y_train[:,i], label='Actual (Train)')
  plt.title(data.columns[i])
  plt.legend()
  plt.tight_layout()
```

Comment: injection rate and proppant concentration is not to be predicted, but rather a completely independent variable for user. Later, we need to modify the forecasting model to predict only the treatment pressure.

```python
from sklearn.metrics import r2_score, mean_absolute_percentage_error

r2_score(y_test[:,-1], y_test_pred[:,-1]), mean_absolute_percentage_error(y_test[:,-1], y_test_pred[:,-1])
```

```python
data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/[2022-2] Kecerdasan Buatan di Industri Minyak dan Gas/for Student (shared)/Chapter6_Frac_Stage_Data.xlsx')
data.drop(['Time'], axis=1, inplace=True)
data.columns
```

```python
data.to_numpy()[0][-1]
```

```python
# create input/output sequences for the RNN
length = 60
X = []  # input array
y = []  # target array
for i in range(len(data) - length - 1):
  X.append(data.to_numpy()[i:i+length])
  y.append(data.to_numpy()[i+length][-1])
```

```python
# convert list to array
X = np.array(X)
y = np.array(y)

X.shape, y.shape
```

```python
# split the data into training and testing sets
split_index = 4400
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN

# Build the RNN model
model = Sequential()
model.add(SimpleRNN(units=60, activation='relu', input_shape=(length, n_features)))
model.add(Dense(units=1))

# Compile and train the model
model.compile(optimizer='adam', loss='mse')
history = model.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
# Make predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# prepare x-axis for testing set
x_min = X_train.shape[0]
x_max = X_train.shape[0] + X_test.shape[0]

plt.figure(figsize=(10,3))
plt.plot(np.arange(x_min, x_max), y_test_pred[:], label='Prediction (Test)')
plt.plot(np.arange(x_min, x_max), y_test[:], label='Actual (Test)')
# plt.plot(y_train_pred[:,-1], label='Prediction (Train)')
# plt.plot(y_train[:], label='Actual (Train)')
plt.title(data.columns[-1])
plt.legend()
plt.tight_layout()
```
