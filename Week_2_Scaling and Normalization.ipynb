# Scaling and Normalization

* scaling dan normalization adalah proces mentrasformasikan suatu data numerik ke dalam nilai lain yang reseprentatif

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ade-irawan/dummy-datasets/main/bikesharing.csv')
data.head()
```

```python
data['windspeed']
```

```python
import seaborn as sns
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

minmiax_scaler = MinMaxScaler()
std_scaler = StandardScaler()
```

```python
data_numeric = data.drop('datetime', axis=1)

minmiax_scaler.fit(data_numeric)
std_scaler.fit(data_numeric)

data_minmax = minmiax_scaler.transform(data_numeric)
data_std = std_scaler.transform(data_numeric)
```

```python
data_minmax= pd.DataFrame(data_minmax, columns= data_numeric.columns)
data_minmax.describe()
```

```python
data_std= pd.DataFrame(data_std, columns= data_numeric.columns)
data_std.describe()
```

```python
fig, ax = plt.subplots(ncols=3)
sns.displot(data['windspeed'], ax=ax[0])
sns.displot(data['humidity'], ax=ax[0])
sns.displot(data['temp'], ax=ax[0])
```
